```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
# Load the Customer dataset

df = pd.read_csv('customer_data_records.csv')
print(df)
```

```python
#Select the Features for clustering
#Store the Features in A Variable

x = df[ [ 'Annual Income (k$)' , 'Spending Score (1-100)'] ]
```

```python
#Apply the K-means Clustering Algorithmn
#Identify the no of clusters to be used(am using two)
#Create a variable to store the KMeans()

kmeans = KMeans(n_clusters=2,random_state=0)
kmeans.fit(x)
```

```python
# Add Cluster labels to the dataset
df['Cluster'] = kmeans.labels_
df
```

```python
#Visualize the Cluster
plt.figure(figsize = (10,8))
sns.scatterplot(data = df , x = 'Annual Income (k$)' , y = 'Spending Score (1-100)' , hue = 'Cluster')
plt.title("Graph showing Customer Clusters based on Income $ Spending ")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score (1-100)")
plt.show()
```
